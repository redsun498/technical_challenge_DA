## Exploratory Data Analysis (EDA)

```python
import pandas as pd 
import numpy as np 
```

```python
df = pd.read_csv('data/measurements.csv')
```

```python
df
```

```python
#Normalizing the names of the columns 
df.columns = df.columns.str.lower().str.replace(' ','_')
df.columns
```

```python
#From intial viewings, distance, consume, temp_inside and refill_liters  should be numerical columns not object
df.info()
```

```python
#Create a list of the potential numerical columns and iterate over them in a loop to see how many values are not numerical
potential_numericals =list(("distance", "consume", "temp_inside","refill_liters",))
for column in potential_numericals:
    print(column, df[column].str.isnumeric().sum())
    print(df[column].unique())
```

```python
#Because of the commas in the previous columns, they are being treated as obj instead of numbers
for column in potential_numericals:
    df[column] = df[column].str.replace(',','.')
```

```python
#Change the data types
for column in potential_numericals:
    df[column] = df[column].astype(float)
```

```python
df.info()
```

```python
#No duplicates
df.duplicated().sum()
```

```python
#refill_liters and refill_gas are nearly entirely null, specials has a high number of nulls as well
#the others are ok  
df.isna().sum()
```
